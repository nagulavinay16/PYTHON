```python
import pandas as pd
df1=pd.DataFrame({
    "city":["mumbai","delhi","hyderabad","chennai"],
    "temperature":[80,60,72,94]
})
df1
```

```python
df2=pd.DataFrame({
    "city":["delhi","mumbai","hyderabad","kolkatta"],
    "winspeed":[80,60,72,84]
})            # to align rows
df2
```

```python
df3=pd.merge(df1,df2,on="city")           # gives only common data(intersection)
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)           # gives all data(union)
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")           # gives all data(left)
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")           # gives all data(right)
df3
```

```python
import pandas as pd
df1=pd.DataFrame({
    "city":["mumbai","delhi","hyderabad","chennai"],
    "temperature":[80,60,72,94],
    "winspeed":[80,60,72,84]
})
df1
```

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
```
