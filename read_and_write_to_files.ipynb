```python
import pandas as pd

def convert_people_cell(cell):
    if cell=="n.a":
        return 'sam walton'
    return cell
def convert_eps_cell(cell):
    if cell=="NOT AVAILABLE":
        return '256'
    return cell

df=pd.read_excel("C:\\Users\\dell\\Downloads\\Desktop\\stock_data.xlsx", converters={
    'people': convert_people_cell,
    'eps':convert_eps_cell
})
df
```

```python
df.to_excel("new.xlsx",sheet_name="stocks",startrow=2,startcol=5,index=False)
```
