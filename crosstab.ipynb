```python
#displays frequency distribution
import pandas as pd
df=pd.read_excel("C:\\Users\\dell\\Downloads\\Desktop\\contingency.xlsx")
df
```

```python
pd.crosstab(df.sex,df.handedness)      #displays frequencies
```

```python
pd.crosstab(df.nationality,[df.handedness,df.sex],margins=True)  #margins displays sum of variables
```

```python
pd.crosstab([df.sex],[df.handedness])  
```

```python
pd.crosstab([df.sex],[df.handedness],normalize='index')  #normalize gives avaerage
```

```python
import numpy as np
pd.crosstab([df.sex],[df.handedness],values=df.age,aggfunc=np.average)  
```
